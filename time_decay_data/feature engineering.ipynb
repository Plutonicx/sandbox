```python
import numpy as np
import pandas as pd
import feather
from tqdm import tqdm
```

```python
tmp_items = feather.read_dataframe('tmp_items.feather')
```

```python
#tmp_items['max_value'] = tmp_items['max_value'].apply(lambda x: float(x), 1)
```

```python
def is_float(x):
    try:
        return(float(x))
    except:
        return(0)
```

```python
tmp_items['max_value'] = tmp_items['max_value'].apply(is_float, 1)
```

```python
tmp_max_items = tmp_items.groupby('property', as_index=False)['max_value'].max()
```

```python
kept_properties = set(tmp_max_items[tmp_max_items['max_value'] != 0]['property'])
```

```python
tmp_items = tmp_items[tmp_items['property'].apply(lambda x: x in kept_properties, 1)]
```

```python
tmp_items_pivot = tmp_items[['itemid','property','max_value']].pivot(values = 'max_value',index = 'itemid',columns = 'property')
```

```python
tmp_items_pivot=tmp_items_pivot.fillna(0)
```

```python
tmp_items_pivot
```

# Normalise columns

```python
for col in tqdm(list(tmp_items_pivot.columns)):
    if 'id' not in col:
        max_val = np.max(tmp_items_pivot[col])
        tmp_items_pivot[col] = tmp_items_pivot[col].apply(lambda x:x / max_val)
```

# Turn IDs into Ints

```python
for col in tqdm(list(tmp_items_pivot.columns)):
    if 'id' in col:
        tmp_items_pivot[col] = tmp_items_pivot[col].apply(lambda x: float(x),1 )
```
