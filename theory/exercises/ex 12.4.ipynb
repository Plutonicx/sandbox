```python
import numpy as np
from plotly import graph_objs as go
from plotly.offline import init_notebook_mode, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
P_2_posterior = np.random.beta(40+1,50-40+1,10000)
P_1_posterior = np.random.beta(30+1,50-30+1,10000)
```

```python
Tau_posterior = P_2_posterior - P_1_posterior
```

```python
data = [
    go.Histogram(x = Tau_posterior,
                 histnorm = "probability density"
                )
]
```

```python
iplot(data)
```

# Get the mles for computation

```python
p_1 = 3/5
p_2 = 4/5
```

```python
grad_psi = np.matrix([[p_2 / (p_1 * (1-p_1)*(1-p_2))],
           [-1*(1-p_1) / ((1-p_2)*p_1*p_2)]])
```

```python
grad_psi
```

```python
J = -1*np.matrix([[p_1*(1-p_1)/50, 0],
    [0, p_2*(1-p_2)/50]])
```

```python
J
```

```python
np.matmul(np.matmul(np.transpose(grad_psi), J),grad_psi)
```

```python
psi_se = np.sqrt(-1*np.matmul(np.matmul(np.transpose(grad_psi), J),grad_psi))
```

```python
float(psi_se)
```

```python
psi_mle = np.log((p_1*(1-p_2))/((1-p_1)*p_2))
```

# using simulation

```python
psi_sim = [ np.log((P_1_posterior[i]*(1-P_2_posterior[i]))/((1-P_1_posterior[i])*P_2_posterior[i])) for i in range(10000) ]
```

```python
data = [
    go.Histogram(x = psi_sim,
                 histnorm = "probability density"
                )
]
```

```python
iplot(data)
```
