```python
import numpy as np
from plotly import graph_objs as go
from plotly.offline import init_notebook_mode, iplot
from scipy.stats import norm
```

```python
init_notebook_mode(connected=True)
```

# Simulating Data as in a)

```python
X = np.random.normal(5,1,100)
```

```python
data = [
    go.Histogram(x = X,
                 histnorm = "probability",
                 autobinx = False,
                 xbins=dict(
                    start=-2,
                    end=9,
                    size=0.5
                )
            )
]
```

```python
iplot(data)
```

```python
mle = np.mean(X)
se = 1/np.sqrt(100)
```

```python
posterior = lambda x: norm(scale = se, loc = mle).pdf(x)
```

```python
posterior(5)
```

```python
pos_sim = np.random.normal(mle, se, 1000)
```

```python
np.max(pos_sim)
```

```python
data = [
    go.Histogram(x = pos_sim,
                 histnorm = "probability density",
                 autobinx = False,
                 xbins=dict(
                    start=-4.7,
                    end=5.5,
                    size=0.05
                ),
                 name = 'posterior hisogram'
            ),
    go.Scatter(
        x = np.linspace(start=4.7,stop=5.5,num=17),
        y = posterior(np.linspace(start=4.7,stop=5.5,num=17)),
        mode = 'line',
        name = 'posterior density'
    )
]
```

```python
iplot(data)
```

```python
theta = np.exp(pos_sim)
```

```python
data = [
    go.Histogram(x = theta)
]
```

```python
iplot(data)
```
