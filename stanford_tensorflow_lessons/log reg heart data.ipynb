```python
X = tf.placeholder(tf.float32, [batch_size, 9])
Y = tf.placeholder(tf.float32, [batch_size, 1])
```

```python
w = tf.Variable(tf.random_normal(shape=[9, 1], stddev=0.05), name='weights')
b = tf.Variable(tf.zeros([1]), name='bias')
```

```python
logits = tf.matmul(X,w) + b
#logits = tf.matrix_determinant(w_sq)*(tf.matmul(X,w) + b)
```

```python
loss = tf.nn.sigmoid_cross_entropy_with_logits(logits=logits,labels=Y)
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(loss)
```

```python
init = tf.global_variables_initializer()
```

# Run GD and predict

```python
with tf.Session() as sess:
    sess.run(init)
    for i in range(n_epochs):
        sess.run([optimizer,loss],feed_dict={X:train_x,Y:train_y})
    
    total_correct_preds = 0
    X_test = tf.placeholder(tf.float32, [len(test_x),9])
    logits_batch = sess.run(tf.matmul(X_test,w) + b,
                                             feed_dict={X_test: test_x})
    preds = tf.nn.sigmoid(logits_batch)
    correct_preds = tf.equal(preds, test_y)
    accuracy = tf.reduce_sum(tf.cast(correct_preds, tf.float32)) 

    total_correct_preds += sess.run(accuracy)
    print("Accuracy {0}".format(total_correct_preds/len(test_x)))
    print(correct_preds.eval())
    if (total_correct_preds/len(test_x) > 0.6):
        print(w.eval())
```
